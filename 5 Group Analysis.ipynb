```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML

xls = pd.ExcelFile(res.excel_path)

def _stars(p):
    if not np.isfinite(p): return ""
    if p < 0.001: return "***"
    if p < 0.01:  return "**"
    if p < 0.05:  return "*"
    if p < 0.10:  return "†"
    return "n.s."

def _fmt_row(label, mean, sem, ci_lo, ci_hi, t, df, p, dz, n, note=""):
    stars = _stars(p)
    return {
        "Тест / contrast": label,
        "Mean ± SEM": f"{mean:.3f} ± {sem:.3f}",
        "95% CI": f"[{ci_lo:.3f}, {ci_hi:.3f}]",
        "t (df)": f"{t:.2f} ({int(df)})" if np.isfinite(t) and np.isfinite(df) else "—",
        "p": f"{p:.4f}" if np.isfinite(p) else "—",
        "dz": f"{dz:.2f}" if np.isfinite(dz) else "—",
        "n": int(n) if np.isfinite(n) else "—",
        "Sig": stars,
        "Примечание": note,
    }

rows = []

# I. PROBE vs RL (in-sample BIC)
try:
    ms = pd.read_excel(res.excel_path, sheet_name="model_stats")
    for ses in ["C", "S"]:
        r = ms[(ms["session"] == ses) & (ms["metric"] == "delta_bic_rl_minus_probe")]
        if r.empty: continue
        r = r.iloc[0]
        rows.append(_fmt_row(
            f"I. ΔΔ BIC (RL−PROBE) | {ses}",
            r["mean"], r["sem"], r["ci95_low"], r["ci95_high"],
            r["t_vs_0"], r["df"], r["p_vs_0"], r["cohen_dz"], r["n"],
            note=">0 → PROBE лучше"
        ))
except Exception as e:
    rows.append({"Тест / contrast": f"I. model_stats (ошибка: {e})"})

# III. Главная интеракция ROI×Block
try:
    inter = pd.read_excel(res.excel_path, sheet_name="stats_interaction_primary")
    for _, r in inter.iterrows():
        reg = str(r.get("regressor", "?"))
        if reg not in ("lambda_actor_post", "Q_actor_post"): continue
        p_col = "p_oriented_vs_0" if "p_oriented_vs_0" in r.index else "p_vs_0"
        t_col = "t_oriented_vs_0" if "t_oriented_vs_0" in r.index else "t_vs_0"
        m_col = "mean_oriented_vm_minus_dl" if "mean_oriented_vm_minus_dl" in r.index else "mean"
        s_col = "sem_oriented_vm_minus_dl" if "sem_oriented_vm_minus_dl" in r.index else "sem"
        ci_lo = r.get("ci95_low_oriented", r.get("ci95_low", np.nan))
        ci_hi = r.get("ci95_high_oriented", r.get("ci95_high", np.nan))
        dz_col = "cohen_dz_oriented" if "cohen_dz_oriented" in r.index else "cohen_dz"
        rows.append(_fmt_row(
            f"III. Интеракция vmPFC>dlPFC | {reg}",
            r[m_col], r[s_col], ci_lo, ci_hi,
            r[t_col], r["df"], r[p_col], r[dz_col], r["n"],
            note="(vmPFC_S−C) − (dlPFC_S−C)"
        ))
except Exception as e:
    rows.append({"Тест / contrast": f"III. interaction (ошибка: {e})"})

# RM-ANOVA ROI×Block
try:
    anova = pd.read_excel(res.excel_path, sheet_name="stats_interaction_primary")
    # попробуем отдельный лист
    try:
        anova2 = pd.read_excel(res.excel_path, sheet_name="stats_anova_roixblock")
        for _, r in anova2.iterrows():
            if "F" not in r or not np.isfinite(float(r["F"])): continue
            eta = r.get("eta_p2", np.nan)
            rows.append({
                "Тест / contrast": f"IIIa. RM-ANOVA ROI×Block | {r.get('regressor','?')}",
                "Mean ± SEM": f"F({r['df_num']:.0f},{r['df_den']:.0f})={float(r['F']):.2f}",
                "95% CI": "—",
                "t (df)": "—",
                "p": f"{float(r['p']):.4f}" if np.isfinite(float(r["p"])) else "—",
                "dz": f"η²p={float(eta):.3f}" if np.isfinite(float(eta)) else "—",
                "n": int(r["n"]) if "n" in r else "—",
                "Sig": _stars(float(r["p"])),
                "Примечание": "2×2 within-subject ANOVA",
            })
    except Exception:
        pass  # лист не найден — ок
except Exception:
    pass

# H1 vmPFC value (Q_actor, S>C)
try:
    h1 = pd.read_excel(res.excel_path, sheet_name="stats_primary_delta_lambda")
    # ищем vmPFC + Q_actor
    r = h1[(h1.get("roi_group", h1.get("session", "")) == "vmPFC") &
           (h1["metric"].str.contains("Q_actor", na=False))] if "metric" in h1.columns else pd.DataFrame()
    if r.empty and "roi_group" in h1.columns:
        r = h1[(h1["roi_group"] == "vmPFC") & (h1["regressor"].str.contains("Q_actor", na=False))]
    if not r.empty:
        r = r.iloc[0]
        p_v = r.get("p_vs_0", r.get("p_delta_vs_0", np.nan))
        t_v = r.get("t_vs_0", r.get("t_delta_vs_0", np.nan))
        m_v = r.get("mean", r.get("mean_delta_S_minus_C", np.nan))
        s_v = r.get("sem", r.get("sem_delta_S_minus_C", np.nan))
        rows.append(_fmt_row(
            "IIIb. H1 vmPFC | Q_actor ΔS−C",
            m_v, s_v, r.get("ci95_low", np.nan), r.get("ci95_high", np.nan),
            t_v, r.get("df", np.nan), p_v, r.get("cohen_dz", np.nan), r.get("n", np.nan),
            note="vmPFC Q (ценность) S > C"
        ))
except Exception as e:
    rows.append({"Тест / contrast": f"IIIb. H1 vmPFC (ошибка: {e})"})

# dlPFC confirmatory
try:
    dlpfc = pd.read_excel(res.excel_path, sheet_name="stats_dlpfc_confirm")
    priority = ["uncertainty_proxy_neg_lambda_actor", "competition_proxy_alt1_minus_actor"]
    shown = 0
    for metric in priority:
        r = dlpfc[dlpfc["metric"].astype(str) == metric] if "metric" in dlpfc.columns else pd.DataFrame()
        if r.empty: continue
        r = r.iloc[0]
        p_v = r.get("p_vs_0", np.nan)
        t_v = r.get("t_vs_0", np.nan)
        rows.append(_fmt_row(
            f"IV. dlPFC | {metric[:35]}",
            r.get("mean", np.nan), r.get("sem", np.nan),
            r.get("ci95_low", np.nan), r.get("ci95_high", np.nan),
            t_v, r.get("df", np.nan), p_v,
            r.get("cohen_dz", np.nan), r.get("n", np.nan),
            note="C > S (uncertainty ↑ with unknowns)"
        ))
        shown += 1
        if shown >= 2: break
except Exception as e:
    rows.append({"Тест / contrast": f"IV. dlPFC confirm (error: {e})"})

# main per-ROI dBeta (S−C)
try:
    prim = pd.read_excel(res.excel_path, sheet_name="stats_primary_delta_lambda")
    key_regs = ["lambda_actor_post"]
    for roi in ["vmPFC", "dlPFC"]:
        for reg in key_regs:
            if "roi_group" in prim.columns and "regressor" in prim.columns:
                r = prim[(prim["roi_group"] == roi) & (prim["regressor"] == reg)]
            else:
                continue
            if r.empty: continue
            r = r.iloc[0]
            rows.append(_fmt_row(
                f"V. {roi} | λ_actor dBeta S−C",
                r.get("mean_delta_S_minus_C", np.nan), r.get("sem_delta_S_minus_C", np.nan),
                r.get("ci95_low", np.nan), r.get("ci95_high", np.nan),
                r.get("t_delta_vs_0", np.nan), r.get("df", np.nan),
                r.get("p_delta_vs_0", np.nan), r.get("cohen_dz", np.nan), r.get("n", np.nan),
            ))
except Exception as e:
    rows.append({"Тест / contrast": f"V. primary delta (error: {e})"})

cols = ["Тест / contrast", "Mean ± SEM", "95% CI", "t (df)", "p", "dz", "n", "Sig", "Примечание"]
summary_df = pd.DataFrame(rows)
for c in cols:
    if c not in summary_df.columns:
        summary_df[c] = ""
summary_df = summary_df[cols].fillna("")

def _highlight(row):
    sig = str(row.get("Sig", ""))
    if sig in ("***", "**", "*"):
        return ["background-color: #d4edda"] * len(row)
    if sig == "†":
        return ["background-color: #fff3cd"] * len(row)
    return [""] * len(row)

print("=" * 80)
print("ФИНАЛЬНЫЙ САММАРИ — ГРУППОВОЙ АНАЛИЗ MEG")
print(f"N субъектов с MEG данными: {res.n_subjects}")
print(f"Excel: {res.excel_path.name}")
print(f"Дата: {pd.Timestamp.now().strftime('%d.%m.%Y')}")
print("=" * 80)
print("Обозначения: *** p<.001  ** p<.01  * p<.05  † p<.10  n.s. р≥.10")
print()

display(
    summary_df.style
    .apply(_highlight, axis=1)
    .set_properties(**{"white-space": "pre-wrap", "font-size": "13px"})
    .set_table_styles([
        {"selector": "th", "props": [("font-weight", "bold"), ("background-color", "#343a40"), ("color", "white")]},
        {"selector": "td", "props": [("border", "1px solid #dee2e6"), ("padding", "6px 10px")]},
        {"selector": "tr:hover td", "props": [("background-color", "#e8f4f8")]},
    ])
    .hide(axis="index")
)

print()
print("─" * 80)
print("─" * 80)
print()
sig_rows = summary_df[summary_df["Sig"].isin(["***", "**", "*"])]
ns_rows  = summary_df[summary_df["Sig"] == "n.s."]
trend_rows = summary_df[summary_df["Sig"] == "†"]
if not sig_rows.empty:
    print("ЗНАЧИМЫЕ ЭФФЕКТЫ:")
    for _, r in sig_rows.iterrows():
        print(f"  ✓ {r['Тест / contrast']}: {r['Mean ± SEM']}, {r['t (df)']}, p={r['p']} {r['Sig']}, dz={r['dz']}")
if not trend_rows.empty:
    print("ТРЕНДЫ (p<.10):")
    for _, r in trend_rows.iterrows():
        print(f"  ~ {r['Тест / contrast']}: {r['Mean ± SEM']}, p={r['p']} {r['Sig']}, dz={r['dz']}")
if not ns_rows.empty:
    print("НЕ ЗНАЧИМО:")
    for _, r in ns_rows.iterrows():
        print(f"  ✗ {r['Тест / contrast']}: p={r['p']}, dz={r['dz']}")

print()
print(f"Excel со всеми таблицами: {res.excel_path}")
print(f"Графики: {res.plots_dir}")
print(f"Полный текстовый отчёт: {res.summary_path}")
```
